```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import numpy as np
import os
import random
import librosa as l
```

```python
data_path = 'vox1_meta.csv'
with open(data_path) as file:
    content = file.readlines()
header = content[:1]
rows = content[1:]
print(header)
print(rows)
```

```python
dataset = pd.read_csv(data_path, delimiter='\t')
data = pd.DataFrame(data=dataset, columns=dataset.columns)
data.head()
```

```python
print("tot speaker:", len(data))
print("tot info:",len(data.columns))
```

```python
# not null value inside the data
if data.isnull().values.any():
    print("find null value\n")
```

```python
# gender distribution
gender = data['Gender']
male=0
female=0
for idx in range(len(gender)):
    if(gender[idx]=='m'):
        male+=1
    elif(gender[idx]=='f'):
        female+=1
    else:
        print("not m or f found")
print("tot m:",male,"\ttot f:",female)
```

```python
my_data = [male, female]
my_labels = 'male','female'
plt.pie(my_data,labels=my_labels,autopct='%1.1f%%')
plt.title('Gender distribution')
plt.axis('equal')
plt.show()
```

```python
dif_nationality = {}
nationality = data['Nationality']
for idx in range(len(nationality)):
    if not nationality[idx] in dif_nationality.keys():
        dif_nationality[nationality[idx]] = 1
    else:
        dif_nationality[nationality[idx]]+=1
dif_nationality
```

```python
my_data = dif_nationality.values()
my_labels = dif_nationality.keys()
figure(figsize=(30, 10))
plt.bar(my_labels, my_data)
plt.title('Nationality distribution')
plt.xlabel('Nationality')
plt.xticks(rotation=45)
plt.ylabel('n speaker')
plt.show()
```

```python
np.random.seed(0)
female_idx = [data['VoxCeleb1 ID'][idx] for idx in range(250) if data['Gender'][idx] == 'f']
print(len(female_idx))
random_f = random.sample(range(len(female_idx)-1), 20) # 20 random numbers about female speaker
print("female:",random_f)
select_f_idx = [female_idx[i] for i in random_f] # select the 20 random idx
for i in range(len(select_f_idx)):
    index = list(data['VoxCeleb1 ID']).index(select_f_idx[i])
    #print("idx:", select_f_idx[i], "gender:", data['Gender'][index])
    if data['Gender'][index] != 'f':
        print("ERROR")
        
np.random.seed(1)
male_idx = [data['VoxCeleb1 ID'][idx] for idx in range(250) if data['Gender'][idx] == 'm']
print(len(male_idx))
random_m = random.sample(range(len(male_idx)-1), 20)
print("male:",random_m)
select_m_idx = [male_idx[i] for i in random_m] # select the 20 random idx
for i in range(len(select_m_idx)):
    index = list(data['VoxCeleb1 ID']).index(select_m_idx[i])
    #print("idx:", select_f_idx[i], "gender:", data['Gender'][index])
    if data['Gender'][index] != 'm':
        print("ERROR")
```

```python
select_f_idx.sort()
select_m_idx.sort()
print(select_f_idx)
print(select_m_idx)
# check that there are not present same id in different gender vector
print(any(np.in1d(select_f_idx, select_m_idx)))
```

```python
path = '../../../../../media/volta/aloba_aloba_yeah/ML/qui/wav/'
print("tot speaker:", len(os.listdir(path)))
sample_rate = 16000
```

```python
mfcc_audio_data_f = []
for i,idx in enumerate(select_f_idx):
    file_name = path + idx +'/'
    subfolder = os.listdir(file_name)
    tot_subfolder = len(subfolder)
    print("spk:",idx,"tot subfolder:", tot_subfolder)
    temp_folder = []
    
    for t,sub_fold in enumerate(subfolder):
        audio_list = os.listdir(file_name + sub_fold +'/')
        tot_audio += len(audio_list)
        temp_store = []
        for x,audio in enumerate(audio_list):
            audio_path = file_name + sub_fold + '/' + audio
            signal, sample_rate = l.load(audio_path, sr=sample_rate)
            mfcc_features = l.feature.mfcc(y=signal, sr=sample_rate, n_mfcc=13, hop_length=int(0.010*sample_rate), 
                                    n_fft=int(0.025*sample_rate))
            temp_store.append(mfcc_features)
            #print("processed spk ",idx, " folder ",sub_fold, " audio ",audio)
        # adesso dovrei applicare la normalizzazione dell'audio
        temp_folder.append(temp_store)
    mfcc_audio_data_f.append(temp_folder)
    print("tot_audio:", len(mfcc_audio_data_f), " - ", len(mfcc_audio_data_f[i]))
```
